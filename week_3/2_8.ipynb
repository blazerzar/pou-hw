```r
library(repr)
options(repr.plot.width = 6, repr.plot.height = 4)
options(repr.plot.res = 200)

library(ggplot2)

set.seed(1)
n <- 1000

original_samples <- runif(n)
die_samples <- ceiling(6 * original_samples)

data.frame(die_samples) |>
  ggplot(aes(x = die_samples)) +
  geom_bar() +
  xlab("Die roll") +
  ylab("Throws") +
  scale_x_continuous(breaks = 1:6)
```
