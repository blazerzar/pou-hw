```r
library(repr)
options(repr.plot.width = 4, repr.plot.height = 4)
options(repr.plot.res = 200)

library(ggplot2)
library(dplyr)
library(tidyr)
library(readr)

set.seed(1)

n <- 1000
```

```r
independent <- data.frame(X = rnorm(n), Y = rnorm(n))
independent |>
  ggplot(aes(x = X, y = Y)) +
  geom_point()
```

```r
options(repr.plot.width = 5, repr.plot.height = 2.5)

dependent <- data.frame(X = rnorm(n)) |>
  mutate(
    Y_1 = rnorm(n, 1 * X, 1),
    Y_0 = rnorm(n, 0 * X, 1),
    "Y_-0.5" = rnorm(n, -0.5 * X, 1),
  ) |>
  pivot_longer(
    cols = starts_with("Y"),
    names_to = "a",
    values_to = "Y",
  ) |>
  mutate(
    a = as.factor(parse_number(a))
  )


dependent |>
  ggplot(aes(x = X, y = Y, color = a)) +
  geom_point(size = 0.3) +
  facet_wrap(~a) +
  coord_equal(ratio = 1)
```

```r
options(repr.plot.width = 5, repr.plot.height = 3)

dependent |>
  ggplot(aes(x = Y, color = a)) +
  geom_density()
```
