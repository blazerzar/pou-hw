```r
library(repr)
options(repr.plot.width = 6, repr.plot.height = 4)
options(repr.plot.res = 200)

library(ggplot2)
library(dplyr)
library(MASS)

set.seed(1)

n <- 1000
l <- matrix(c(sqrt(2), 1.2 / sqrt(2), 0, sqrt(0.28)), nrow = 2)
mu <- c(1.5, -1)
```

```r
samples_univariate <- t(l %*% rbind(rnorm(n, 0, 1), rnorm(n, 0, 1)) + mu)
samples_multivariate <- mvrnorm(n, mu, l %*% t(l))
```

```r
data.frame(rbind(samples_univariate, samples_multivariate)) |>
  bind_cols(distribution = c(rep("univariate", n), rep("multivariate", n))) |>
  ggplot(aes(x = X1, y = X2, color = distribution)) +
  geom_point(alpha = 0.5)
```
