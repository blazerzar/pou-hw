```r
library(repr)
options(repr.plot.width = 6, repr.plot.height = 4)
options(repr.plot.res = 200)

library(ggplot2)
library(reshape2)

set.seed(1)

mu <- 0.4
sigma <- sqrt(0.25)

n <- 10000
samples <- rnorm(n, mu, sigma)
```

```r
# a) E[2X + 4]
print(c(2 * mu + 4, mean(2 * samples + 4)))

# b) E[X^2]
print(c(sigma^2 + mu^2, mean(samples^2)))

# c) E[exp(X)]
print(c(exp((sigma^2 + 2 * mu) / 2), mean(exp(samples))))
```

```r
data.frame(
  X1 = samples, X2 = 2 * samples + 4, X3 = samples^2, X4 = exp(samples)
) |>
  melt(id.vars = c()) |>
  ggplot(aes(x = value, color = variable)) +
  geom_density() +
  scale_color_discrete(
    name = "Distribution",
    labels = c("X", "2X + 4", "X^2", "e^X")
  )
```
