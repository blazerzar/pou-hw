```r
library(repr)
options(repr.plot.width = 5, repr.plot.height = 4)
options(repr.plot.res = 200)

library(ggplot2)
library(dplyr)

set.seed(1)

n <- 10000
```

```r
x_sample <- rnbinom(n, 5, 1 - 0.4)
y_sample <- rnbinom(n, 3, 1 - 0.4)
z_sample <- rnbinom(n, 8, 1 - 0.4)

bind_rows(
  data.frame(value = x_sample, distribution = "X"),
  data.frame(value = y_sample, distribution = "Y"),
  data.frame(value = x_sample + y_sample, distribution = "X + Y"),
  data.frame(value = z_sample, distribution = "Z"),
) |>
  ggplot(aes(x = value)) +
  geom_bar() +
  facet_wrap(~ factor(distribution, levels = c("X", "Y", "X + Y", "Z")))
```
