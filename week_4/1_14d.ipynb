```r
library(repr)
options(repr.plot.width = 6, repr.plot.height = 4)
options(repr.plot.res = 200)

library(ggplot2)
library(tidyr)

set.seed(1)

n_samples <- 1000
lambda <- 10
```

```r
samples <- rpois(n_samples, lambda)
counts <- table(samples)
```

```r
data.frame(
  k = c(as.numeric(names(counts)), 0:22),
  frequency = c(as.vector(counts) / n_samples, dpois(0:22, lambda)),
  type = rep(c("sampled", "theoretical"), c(length(counts), 23))
) |>
  ggplot(aes(x = k, y = frequency, fill = type)) +
  geom_bar(position = "dodge", stat = "identity") +
  scale_x_continuous(breaks = seq(0, 22, 2))
```
