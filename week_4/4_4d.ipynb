```r
library(repr)
options(repr.plot.width = 6, repr.plot.height = 4)
options(repr.plot.res = 200)

library(ggplot2)
library(tidyr)

set.seed(1)

n_samples <- 1000
n <- 10
p <- 0.5
```

```r
bernoulli_samples <- colSums(matrix(rbinom(n_samples * n, 1, p), nrow = n))
binomial_samples <- rbinom(n_samples, n, p)
```

```r
data.frame(bernoulli_samples, binomial_samples) |>
  pivot_longer(cols = c(1, 2), values_to = "k", names_to = "distribution") |>
  ggplot(aes(x = k, fill = distribution)) +
  geom_bar(position = "dodge") +
  scale_x_continuous(breaks = 0:n)
```
