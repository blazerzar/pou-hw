```r
library(repr)
options(repr.plot.width = 6, repr.plot.height = 4)
options(repr.plot.res = 200)

library(ggplot2)
library(tidyr)

set.seed(1)
```

```r
normal_mean <- vector(length = 1000)
normal_median <- vector(length = 1000)
cauchy_mean <- vector(length = 1000)
cauchy_median <- vector(length = 1000)

for (n in 1:1000) {
  normal <- rnorm(n, 0, 1)
  normal_mean[n] <- mean(normal)
  normal_median[n] <- median(normal)

  cauchy <- rcauchy(n, 0, 1)
  cauchy_mean[n] <- mean(cauchy)
  cauchy_median[n] <- median(cauchy)
}
```

```r
data.frame(
  n = 1:1000, normal_mean, normal_median, cauchy_mean, cauchy_median
) |>
  pivot_longer(
    cols = 2:5, names_to = c("distribution", "summary"), names_sep = "_"
  ) |>
  ggplot(aes(x = n, y = value, color = distribution)) +
  geom_line() +
  facet_wrap(~ summary + distribution, scales = "free") +
  theme(legend.position = "none")
```
