This is a mixed random variable.


```r
library(repr)
options(repr.plot.width = 6, repr.plot.height = 4)
options(repr.plot.res = 200)

library(ggplot2)

p <- 0.3

cdf <- function(x, p) {
  return(
    ifelse(x < 0, 0,
      ifelse(x < 1, x^2 / 2,
        ifelse(x < 2, 0, (1 - p) / 2) + 0.5 + p / 2
      )
    )
  )
}

ggplot() +
  geom_function(
    fun = cdf, args = list(p = p), color = "blue", n = 1000, xlim = c(-1, 3)
  ) +
  geom_hline(yintercept = 1, color = "black")
```
