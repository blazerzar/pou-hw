```r
library(repr)
options(repr.plot.width = 6, repr.plot.height = 4)
options(repr.plot.res = 200)

library(ggplot2)
library(reshape2)

set.seed(1)

n <- c(1, 2, 5, 10, 100, 1000)
```

```r
x <- seq(-5, 5, 0.01)

data.frame(x = x) |>
  cbind(
    setNames(data.frame(sapply(n, function(i) pnorm(x, 0, 1 / i))), n)
  ) |>
  melt(id.vars = "x", variable.name = "n", value.name = "F_Xn") |>
  ggplot(aes(x = x, y = F_Xn, color = n)) +
  geom_line()
```
